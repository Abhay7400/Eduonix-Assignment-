```python
class ATM:

   

    def __init__(self, pin, initial_balance=10000):
    

        self.pin = pin
        self.balance = initial_balance

    def verify_pin(self, entered_pin):
        
   

        return self.pin == entered_pin

    def check_balance(self):
     

        print(f"Your current balance is: ₹{self.balance:.2f}")

    def withdraw(self, amount):
       

        # Simulate a withdrawal fee of 2.5%
        withdrawal_fee = amount * 0.025
        total_amount = amount + withdrawal_fee

        if total_amount > self.balance:
            print("Insufficient funds. Maximum withdrawal amount: ₹{:.2f}".format(self.balance))
            return False

        self.balance -= total_amount
        print(f"₹{amount:.2f} withdrawn. Your remaining balance is: ₹{self.balance:.2f}")
        return True

    def deposit(self, amount):
     
      

        self.balance += amount
        print(f"₹{amount:.2f} deposited. Your new balance is: ₹{self.balance:.2f}")

    def run(self):
       

        while True:
            print("\n***** Welcome To ATM Machine Simulator *****\n")

            entered_pin = int(input("Enter Your Pin: "))

            if self.verify_pin(entered_pin):
                while True:
                    print("\nOptions you can Exercise are:")
                    print("1) Balance")
                    print("2) Withdraw")
                    print("3) Deposit")
                    print("4) Exit")

                    choice = int(input("\nSelect Your Transaction from the above options: "))

                    if choice == 1:
                        self.check_balance()
                    elif choice == 2:
                        withdrawal_amount = float(input("Enter Amount to Withdraw: "))
                        self.withdraw(withdrawal_amount)
                    elif choice == 3:
                        deposit_amount = float(input("Enter Amount to Deposit: "))
                        self.deposit(deposit_amount)
                    elif choice == 4:
                        print("\nThank you for using the ATM. Have a nice day!")
                        break
                    else:
                        print("Invalid choice. Please try again.")
            else:
                print("Invalid PIN. Please try again.")

            # Prompt to continue or exit
            continue_choice = input("\nDo you want to continue (y/n)? ")
            if continue_choice.lower() != 'y':
                break


if __name__ == "__main__":
    
    atm = ATM(65566)

   
    atm.run()
```
